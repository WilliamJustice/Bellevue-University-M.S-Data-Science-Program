## Begin senitment analysis on the customer review comments and the host description

#### Customer Reviews Senitment Analysis

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

```python
reviews.dropna(inplace=True) # There were some NA's that were not allowing nltk to process
```

```python
# Create a new column in the dataframe to show the polarity ratings for each comment
reviews['PolarityRating'] = reviews['comments'].apply(lambda review: sid.polarity_scores(review))
```

```python
# Separate the compound score. The Compound score is a metric that calculates the sum of all the lexicon ratings which have been normalized between -1(most extreme negative) and +1 (most extreme positive).
reviews['Compound Score']  = reviews['PolarityRating'].apply(lambda score_dict: score_dict['compound'])
```

```python
reviews.head()
```

```python
reviews.to_csv('ReviewsCompunds2.csv')
```

```python
# Create a new column in the dataframe to show the polarity ratings for each comment
listing['PolarityRating'] = listing['description'].apply(lambda review: sid.polarity_scores(review))
```
